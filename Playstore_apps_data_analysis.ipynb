# Importing Liabrearies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Top-Apps-in-Google-Play.csv")
```

```python
df.head()
```

```python
df.describe()
```

# Information about data

```python
df.info()
```

# Checking for nan/missing values

```python
df.isnull().sum()
```

# Types of categories of apps exists on playstore

```python

plt.figure(figsize=(10,8))
sns.countplot(data=df,y="Category")
plt.title("Categorywise Distribution")
```

# Types of developer id's on playstore

```python
df["Developer Id"].value_counts()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x="Developer Id",data=df)
plt.title("Developer ID's")
plt.ylabel("Value counts of id's")
plt.xlabel("Developer Name")
```

# Categoriwise content rating

```python
df["Content Rating"].value_counts().plot(kind="bar")
plt.grid(True)
plt.xlabel("Categories")
plt.ylabel("Value counts")
```

# Checking for add supported apps or not

```python
plt.figure(figsize=(10,10))
x = ["Not Supported","Supported"]
df["Ad Supported"].value_counts().plot(kind="pie",autopct="%.2f%%",shadow=True,explode=(0,0.1),labels=x)
plt.title("Add Supported apps distribution")
```

# Distribution of paid and free apps on playstore

```python

df["In App Purchases"].value_counts().plot(kind="bar")
plt.legend()
plt.title("App Paid or free distribution")
```

```python
plt.figure(figsize=(8,10))
names = ["Unpaid","paid"]
df["In App Purchases"].value_counts().plot(kind="pie",labels=names,autopct="%.2f%%",shadow=True,explode=(0.1,0.0))


plt.title("App Paid or free distribution")
```
